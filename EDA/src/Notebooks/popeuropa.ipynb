```python
# Importo librer√≠as necesarias
import numpy as np, pandas as pd
```

```python
# Importo y leo el dataframe
europa = pd.read_csv('data\\paises.tsv', sep='\t')
europa
```

```python
# Elimino columnas que a primera vista s√© que no necesito
europa = europa.drop(columns= ['2011 ', '2012 '])
```

```python
# Renombro la columna del pa√≠s
europa.rename(columns={'indic_de,geo\\time' : 'Country'}, inplace = True)
```

```python
# Corrijo los valores de pa√≠ses.
europa['Country'] = europa['Country'].apply(lambda valor: valor[-2:])
```

```python
siglas = {'AT': 'Austria',
        'BA': 'Bosnia y Herzegovina',
        'BE': 'B√©lgica',
        'CH': 'Suiza',
        'CY': 'Chipre',
        'CZ': 'Rep√∫blica Checa',
        'DE': 'Alemania',
        'DK': 'Dinamarca',
        'EE': 'Estonia',
        'ES': 'Espa√±a',
        'FI': 'Finlandia',
        'FR': 'Francia',
        'UK': 'Reino Unido',
        'EL': 'Grecia',
        'HR': 'Croacia',
        'IS': 'Islandia',
        'IT': 'Italia',
        'LT': 'Lituania',
        'LV': 'Latvia',
        'MT': 'Malta',
        'NL': 'Pa√≠ses Bajos',
        'NO': 'Noruega',
        'PL': 'Polonia',
        'PT': 'Portugal',
        'RO': 'Rumania',
        'RS': 'Serbia',
        'SE': 'Suecia',
        'SI': 'Eslovenia',
        'SK': 'Eslovaquia',
        'TR': 'Turqu√≠a'}
```

```python
# Cambio de siglas a nombres completos
europa['Country'] = europa['Country'].apply(lambda valor: siglas.get(valor))
```

```python
# Elimino pa√≠ses que no coinciden con el dataframe de aguas residuales
europa = europa.dropna(subset=['Country'])
```

```python
# Indico los pa√≠ses como √≠ndice.
europa.set_index(europa.columns[0], inplace= True)
```

```python
# Elimino las letras sobrantes de los datos.
import re

for columna in europa.columns:
    europa[columna] = europa[columna].replace(to_replace=r'[A-Za-z]+$', value='', regex=True)
```

```python
# Reemplazo el valor ":" a NaN
europa = europa.replace(to_replace=r'^\s*:\s*$', value=np.nan, regex=True)
```

```python
europa = europa.apply(pd.to_numeric)
```

```python
# Leo un dataframe diferente  que contiene algunos de los datos faltantes, para complementar el primero.
missing = pd.read_csv('data/missing.csv')
```
